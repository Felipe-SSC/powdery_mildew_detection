### Modelo MobileNet v3 

```python
import torch
from torchvision.models import mobilenet_v3_large

# instanciar arquitectura sin pesos (None)
model = mobilenet_v3_large(weights=None)

# modificar el clasificador final según el training del modelo
model.classifier[3] = torch.nn.Linear(model.classifier[3].in_features, 2)

# cargar los pesos (.pth)
model.load_state_dict(torch.load('mobilenetv3_large_predictor.pth'))

# poner el modelo en modo evaluacion
model.eval()
```

### Modelo resnet50

```python
import torch
from torchvision.models import resnet50

# instanciar arquitectura sin pesos (None)
model = resnet50(weights=None)

# modificar el clasificador final según el training del modelo
model.fc = torch.nn.Linear(2048, 2)

# cargar los pesos (.pth)
model.load_state_dict(torch.load('resnet50_powdery_mildrew.pth'))

# poner el modelo en modo evaluacion
model.eval()
```
